##  Разделение классов


```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_json("../data/raw/dataset_clean.jsonl", lines=True)

val_and_train, test = train_test_split(df, test_size=0.33, train_size=0.67, stratify=df['label'])

val, train = train_test_split(val_and_train, test_size=0.5, train_size=0.5, stratify=val_and_train['label'])


```

## Соотношение классов


```python
val['label'].value_counts(normalize=True)

```

```python
test['label'].value_counts(normalize=True)
```

```python
train['label'].value_counts(normalize=True)
```

## Сохранение в `.jsonl`

```python
train.to_json("../data/processed/train.jsonl", lines=True, orient='records', force_ascii=False)
val.to_json("../data/processed/val.jsonl", lines=True, orient='records', force_ascii=False)
test.to_json("../data/processed/test.jsonl", lines=True, orient='records', force_ascii=False)
```
